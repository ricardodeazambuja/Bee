##Measuring the difference between the generated curves
$MeanDistance=\frac{1}{N} \sum \limits_{n=1}^{N}  \lVert P_{Ideal}(n)-P_{Test}(n) \rVert$

$TotalLength=\sum \limits_{n=1}^{N}  \lVert P_{Ideal}(n)-P_{Ideal}(n-1) \rVert$

$METRIC=\frac{MeanDistance}{TotalLength}$

```python
# set_number = 0

# sim_set_array=numpy.array(["set_00","set_01","set_02","set_03","set_04","set_05","set_06","set_07","set_08","set_09","set_xx","set_yy"])
sim_set_array=numpy.array(["set_00","set_01","set_02","set_03","set_06","set_07","set_08","set_xx","set_yy"])
sim_set_names=["Set A", "Set B", "Set C", "Set D", "Set E", "Set F", "Set G", "Set H", "Set I"]
# sim_set = sim_set_array[set_number]

base_dir = "2DofArm_simulation_data"
```

```python
# This is for the trajectory
# METRIC[noiselevel,decimationlevel,trial]
# The number 4 is related with the number of trajectories!
METRIC_ALL=numpy.zeros((12, 4, 3, 3, 50, 1),dtype=numpy.float)
for set_number in range(len(sim_set_array)):
    sim_set = sim_set_array[set_number]
    for tji in range(4):
        METRIC=slf.load_from_file("./"+base_dir+"/"+sim_set+"/metric_trajectory_trajectory_"+str(tji+1)+".pickles")
        METRIC_ALL[set_number,tji]=numpy.copy(METRIC)[:,:,:,:] # the :5 is used because the set_00 
                                                                # has 50 trials instead of 5
```

```python
# All trajectories individually
noiselevel=0
decimationlevel=0
    
subplti=1
plt.figure(figsize =(20,20))
for tji in range(1,5):
    plt.subplot(2,2,subplti)
    subplti+=1
    plt.errorbar(numpy.arange(len(sim_set_array)),\
                 [METRIC_ALL[set_i,tji-1,noiselevel,decimationlevel,:].mean() for set_i in range(len(sim_set_array))],\
                 [METRIC_ALL[set_i,tji-1,noiselevel,decimationlevel,:].std()/numpy.sqrt(len(METRIC_ALL[set_i,tji-1,noiselevel,:,:])) for set_i in range(len(sim_set_array))],\
                 linestyle='None', marker='s')
    plt.plot(numpy.arange(len(sim_set_array)),[METRIC_ALL[set_i,tji-1,noiselevel,decimationlevel,:].mean() for set_i in range(len(sim_set_array))],'-',label="N:"+['1x','2x','10x'][noiselevel] +" D:"+['0%','1%','10%'][decimationlevel])

    plt.title("Trajectory: "+str(tji)+" - noise level:"+['1x','2x','10x'][noiselevel]+" - decimation level:"+['0%','1%','10%'][decimationlevel])
    plt.ylabel("Normalized Error (trajectory "+str(tji) +")")
    plt.xticks(numpy.arange(len(sim_set_array)), sim_set_names, rotation='vertical')
    plt.xlim(-1,len(sim_set_array))
    plt.ylim(0,0.5)
#     plt.grid()
    plt.yticks(numpy.linspace(0,0.5,num=20))
plt.subplots_adjust(left=0, bottom=.1, right=1, top=1,wspace=.2, hspace=.5)
plt.savefig("./"+base_dir+"/NormalizedError_individual_traj.pdf", bbox_inches='tight',pad_inches=1)
plt.show()        
```
